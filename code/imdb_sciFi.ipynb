```python
# import required packages
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import numpy as np
import re
from time import sleep
from time import time
from random import randint
pd.set_option('display.width', 300, 'display.max_rows', 300, 'max_colwidth',200)
```

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}

# Lists to store the scraped data in
names = []
years = []
imdb_ratings = []
votes = []

# set the startNumber (5 pages, 200 movies)
start_nums = []
for i in range(0, 5):
    start_nums.append(str(i*50+1))

# For every page in the interval
for start_num in start_nums:
    # Make a get request
    url = 'https://www.imdb.com/search/title/?title_type=feature&genres=sci-fi&sort=num_votes,desc&start=' + start_num + '&ref_=adv_nxt'
    # print(url)
    response = requests.get(url, headers = headers)
    # print(response.status_code)

    # Pause the loop
    sleep(randint(3,10))

    # Parse the content of the request with BeautifulSoup
    html_soup = BeautifulSoup(response.content, 'html.parser')

    # Select all the 50 movie containers from a single page
    movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')

    # For every movie of these 50
    # Extract data from individual movie container
    for container in movie_containers:
        # The name
        name = container.select('.lister-item-header a')[0].get_text()
        names.append(name)
        # The year
        yearTemp = container.find('span', class_ = 'lister-item-year').get_text()
        # print(yearTemp, '---')
        # Regular Experssion to Reg the year
        # year = re.search(r'[(](\d{4,})[)]', yearTemp)
        year = re.search(r'(\d{4,})', yearTemp)
        year = year.group()
        # print(year)
        years.append(year)
        # The IMDB rating
        imdb = float(container.find('strong').get_text())
        imdb_ratings.append(imdb)
        # imdbTemp = container.find('strong')
        # if imdbTemp is not None:
        # imdb = float(imdbTemp.get_text())
        # else:
            # imdb = 0
        # The number of votes
        vote = container.find('span', attrs={'name':'nv'})['data-value']
        votes.append(int(vote))
        # if voteTemp is not None:
            # vote = voteTemp['data-value']
        # else:
        #     vote = 0
            # votes.append(int(vote))
sciFiMovie_ratings = {
    'movie': names,
    'year': years,
    'imdb': imdb_ratings,
    'votes': votes
}

# sciFiMovie_ratings_df = pd.DataFrame.from_dict(sciFiMovie_ratings, orient='index')
# df = sciFiMovie_ratings_df.transpose()

sciFiMovie_ratings_df = pd.DataFrame(sciFiMovie_ratings)

print(sciFiMovie_ratings_df.info())
```

```python
# scrap summary and storyline from single movie pages
# clean the data to get the format what we need

headers = {"Accept-Language": "en-US, en;q=0.5"}

#list to store the scrap data
summaries = []
storylines = []

# set the startNumber (5 pages, 200 movies)
start_nums = []
for i in range(0, 5):
    start_nums.append(str(i*50+1))

domain = 'https://www.imdb.com'
# For every page in the interval
for start_num in start_nums:
    # Make a get request
    url = 'https://www.imdb.com/search/title/?title_type=feature&genres=sci-fi&sort=num_votes,desc&start=' + start_num + '&ref_=adv_nxt'
    print(url)
    response = requests.get(url, headers = headers)
    # print(response.status_code)

    # Pause the loop
    sleep(randint(3,10))

    # Parse the content of the request with BeautifulSoup
    html_soup = BeautifulSoup(response.content, 'html.parser')

    # Select all the 50 movie containers from a single page
    movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')

    # Extract data from individual movie container
    for container in movie_containers:
        sub_url = container.select('.lister-item-header a')[0]['href']
        # url of movie page
        movie_url = domain + sub_url
        movie_html = requests.get(movie_url, headers = headers)
        # print(movie_html)
        # pause the loop
        sleep(randint(3,10))
        # extract summary information
        movie_soup = BeautifulSoup(movie_html.content, 'html.parser') # parser of individual movie page
        summaryTemp = movie_soup.find('div', class_='summary_text').get_text()
        summary = summaryTemp.replace('\n', '')
        summary = summary.strip() # delete all the whitespace
        summaries.append(summary)
        # extract storyline information
        storylineDiv = movie_soup.find('div', class_='inline canwrap')
        storylineSpan = storylineDiv.find('span').get_text()
        storyline = storylineSpan.replace('\n', '')
        storyline = storyline.strip()
        storylines.append(storyline)


sciFiMovie_contents = {
    'summary': summaries,
    'storyline': storylines
}

sciFiMovie_contents_df = pd.DataFrame(sciFiMovie_contents)

print(sciFiMovie_contents_df.info())
```

```python
# Concatenate two dataframes
result = pd.concat([sciFiMovie_ratings_df, sciFiMovie_contents_df], axis=1)
result
```
